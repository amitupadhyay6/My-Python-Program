```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#We need subplot to plot 2 graph, else both graph will be merge in one.
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.2)

#plt.subplott(111)
plt.plot(t1, f(t1), t2, f(t2))

#plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2))
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.2)

plt.subplot(221)
plt.title("1st Graph")
plt.plot(t1, f(t1), t2, f(t2))

plt.subplot(222)
plt.title("2nd Graph")
plt.plot(t2, np.cos(2 * np.pi * t2))
plt.show()
```
